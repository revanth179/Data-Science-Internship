```python
class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        for i in range(1,len(nums)):
            nums[i]+=nums[i-1]
        return nums
```

```python
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        a = []
        index = 0
        for i in range(n):
            a.append(nums[i])
            a.append(nums[i + n])
        return a
```

```python
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        l = max(candies)
        m = []
        for i in candies:
            if i + extraCandies >= l:
                m.append(True)
            else:
                m.append(False)
        return m
```
